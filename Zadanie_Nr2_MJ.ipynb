#punkt b

```python
# imprt stack
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
#
#DataFrame
iris = load_iris() #wczytaj klase iris()
feature_names = iris['feature_names']  # 4 pierwsze kolumny cech z iris
df = pd.DataFrame(iris['data'], columns=feature_names) #Dataframe
df['target'] = iris['target'] #etykiety 0/1/2 jako target
#
#kopia przed normalizacją
df_norm = df.copy(deep=True)
#
#
scaler1 = MinMaxScaler(feature_range=(0, 1)) #MinMaxScaler - zakres [0,1]
cols_to_scale = feature_names  # pierwsze 4 kolumny, tylko cechy
df_norm[cols_to_scale] = scaler1.fit_transform(df_norm[cols_to_scale])
#

#sprawdzczy kolumny cech są w zakresie [0,1]
mins = df_norm[cols_to_scale].min().round(6)
maxs = df_norm[cols_to_scale].max().round(6)
print("MIN po normalizacji:\n", mins, "\n")
print("MAX po normalizacji:\n", maxs, "\n")
#
# sprawdz czy jest ok, jeśli nie wyswietla bład
assert (mins >= 0).all() and (maxs <= 1).all(), "Skalowanie nie jest w zakresie 0-1"
#
#podsumuj znormalizowanego zbioru
print("\n=== SUMMARY===")
print(df_norm[cols_to_scale].describe().T)
#
# porównanie przed/po dla pierwszych 5 wierszy
print("\nPRZED -> PO:")
print(
    pd.concat(
        [df[cols_to_scale].head().add_suffix(" [orig]"),
         df_norm[cols_to_scale].head().add_suffix(" [min-max]")],
        axis=1
    )
)
```

#punkt c

stwórz kopię zbioru iris i znormalizuj w nim dane z kolumn 1-4 metodą standaryzacji Zscore
korzystając z funkcji preProcess i predict (argument method=c("center", "scale") ) (Uwaga !:
proszę pamiętać o odpowiednim wywołaniu funkcji predict po preProcess, jak to jest
pokazane w sekcji Examples na końcu strony z opisem tych funkcji). Sprawdź, czy wartości
podanych kolumn zostały odpowiednio zmodyfikowane. Wypisz podsumowanie
znormalizowanego zbioru (funkcja summary).

```python
# import z-score
from sklearn.preprocessing import StandardScaler  # Z-score
#
# kopia
df_z = df.copy(deep=True)
# standaryzacja Z-score
scaler2 = StandardScaler(with_mean=True, with_std=True)  # Z-score
cols_to_scale = feature_names  # tylko cechy (pierwsze 4 kolumny)
df_z[cols_to_scale] = scaler2.fit_transform(df_z[cols_to_scale])
#
# weryfikacja: średnia ~ 0, std ~ 1 (czyli wartości jak w dokumentacji [-1,1]68%, [-2,2]-95%, [-3,3]-99,7% dla rozkł. gaussa)
means = df_z[cols_to_scale].mean().round(6)
stds  = df_z[cols_to_scale].apply(lambda s: np.std(s.values, ddof=0)).round(6)
#
print("AVG po standaryzacji (≈ 0):\n", means, "\n")
print("STD po standaryzacji (≈ 1):\n", stds, "\n")
#
# sprawdz czy jest ok, jeśli nie wyswietla bład
assert np.allclose(df_z[cols_to_scale].mean().values, 0.0, atol=1e-8), "Średnie nie są ~0!"
assert np.allclose(
    df_z[cols_to_scale].apply(lambda s: np.std(s.values, ddof=0)).values,
    1.0, atol=1e-8
), "Odchylenia standardowe nie są ~1!"
#
#wyswietl podsumowanie
print("\n=== SUMMARY (Z-score, cechy) ===")
print(df_z[cols_to_scale].describe().T)
#
#porównanie przed/po dla pierwszych 5 wierszy
print("\nPRZED -> PO:")
print(
    pd.concat(
        [df[cols_to_scale].head().add_suffix(" [orig]"),
         df_z[cols_to_scale].head().add_suffix(" [z-score]")],
        axis=1
    )
)
```

porównanie punkt b i c w jednym kodzie

```python
#kopie jeszcze raz do porównania
df_minmax = df.copy()
df_zscore = df.copy()
cols = feature_names

#Min–Max [0,1]
scaler1 = MinMaxScaler(feature_range=(0, 1))
df_minmax[cols] = scaler1.fit_transform(df_minmax[cols])
#
#Z-score
scaler2 = StandardScaler()
df_zscore[cols] = scaler2.fit_transform(df_zscore[cols])

#Porównanie
print("\n=== PORÓWNANIE: MIN–MAX vs Z–SCORE ===")
print(pd.concat([
    df_minmax.head().add_suffix(" [min–max]"),
    df_zscore.head().add_suffix(" [z–score]")
], axis=1))
#
#Zakresy
print("\nZakresy wartości:")
print(pd.DataFrame({
    "min–max_min": df_minmax[cols].min(),
    "min–max_max": df_minmax[cols].max(),
    "z–score_min": df_zscore[cols].min().round(2),
    "z–score_max": df_zscore[cols].max().round(2)
}))
```
