W oparciu o artykuł Dealing with Missing Data using R przeprowadź za pomocą pakietu Hmisc
imputację danych (sztuczne wstawienie odpowiednich wartości do tabeli) w zmodyfikowanym
zbiorze iris (iris.mis) zawierającym brakujące wartości.

Poczytałem trochę trochę w R jak wyglądają dane i wrzucam generację w Python

```python
# normal stack import
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris() #wczytuje dane z klasy looad_iris()
X = iris['data']          # i przypisuję cechy do X
y = iris['target']        # gatunki do Y - target
feature_names = iris['feature_names'] #etykiety kolumn - dane wejściowe
species_names = iris['target_names'] #przyporządkowaanie etykiet dla targetu

iris3 = np.empty((50, 4, 3)) #tworzę pustą tablicę 50 wierszy, 4 cechy, 3 typy gatunków
for i in range(3):
    iris3[:, :, i] = X[y == i]

print("Kształt tablicy iris3:", iris3.shape)

np.random.seed(42)  # daje losowe seed(42) braki dla powtarzalności
missing_ratio = 0.1  # 10% braków - ustalam 10% braków

# ile elementów wyzerować
n_total = iris3.size #łączna liczba elementów
n_missing = int(missing_ratio * n_total) #obliczenie ile elementów wyzerować

#wydukuj podsumę
print("Łączna liczba elementów (n_total):", n_total)
print("Liczba elementów do wyzerowania (n_missing):", n_missing)


# losuje indeksy i wstawiam NaN generując puste dane
indices = np.unravel_index(
    np.random.choice(n_total, n_missing, replace=False),
    iris3.shape
)

iris3[indices] = np.nan

print("Liczba brakujących wartości po wstawieniu:", np.isnan(iris3).sum()) #suma pustych powinna się zgadzać



# Przekstzłcam do DataFrame dla łatwiejszego przeglądu
iris_df = pd.DataFrame(
    data=np.vstack([iris3[:, :, i] for i in range(3)]),
    columns=feature_names
)
iris_df['species'] = np.repeat(species_names, 50)

# Podsumowanie braków
print("\nBraki w kolumnach:")
print(iris_df.isna().sum())

# Podgląd fragmentu danych
print("\nPrzykładowe dane z brakami:")
print(iris_df.head(10))
```

```python
# średnia z kolumny 'sepal length (cm)' pomijac  NaN
mean_sepal_length = iris_df['sepal length (cm)'].mean()

# nowa kolumna z wartością średnią
iris_df['sepal length_mean (cm)'] = iris_df['sepal length (cm)'].fillna(mean_sepal_length)

#wypisz
print("Średnia wartość (bez NaN):", mean_sepal_length)
iris_df[['sepal length (cm)', 'sepal length_mean (cm)']].head(10)
```
